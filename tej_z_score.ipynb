```python
A_2324 = (df1.loc['2324 Compal','Total current assets'] - df1.loc['2324 Compal','Total current liabilities'])/df1.loc['2324 Compal','Total assets']
```

```python
A_2324
```

```python
X1 ='Total current assets'
X2 ='Total current liabilities'
X3 ='Total assets'
X4 ='Total retained earnings'
X5 ='Earnings before interest and tax'
X6 ='Total equity'
X7 ='Total liabilities'
X8 ='Total operating revenue'
Compel ='2324 Compal'
ACER ='2353 ACER'
```

```python
A_compel = (df1.loc[Compel,X1] - df1.loc[Compel,X2])/df1.loc[Compel, X3]
```

```python

B_compel = df1.loc[Compel,X4]/df1.loc[Compel, X3]
```

```python
C_compel = df2.loc[Compel,X5]/df1.loc[Compel, X3]
```

```python
D_compel = df1.loc[Compel,X6]/df1.loc[Compel, X7] #調整分子項定義，採用股東權益帳面價值
```

```python
E_compel = df2.loc[Compel,X8]/df1.loc[Compel, X3]
```

```python

Z_score_compel = 1.2*A_compel + 1.4*B_compel + 3.3*C_compel + 0.6*D_compel +1.0*E_compel
```

```python
Z_score_compel, A_compel, B_compel, C_compel, D_compel, E_compel
```

```python
A_ACER = (df1.loc[ACER,X1] - df1.loc[ACER,X2])/df1.loc[ACER, X3]


B_ACER = df1.loc[ACER,X4]/df1.loc[ACER, X3]


C_ACER = df2.loc[ACER,X5]/df1.loc[ACER, X3]


D_ACER = df1.loc[ACER,X6]/df1.loc[ACER, X7]


E_ACER = df2.loc[ACER,X8]/df1.loc[ACER, X3]


Z_score_ACER = 1.2*A_ACER + 1.4*B_ACER + 3.3*C_ACER + 0.6*D_ACER +1.0*E_ACER


Z_score_ACER, A_ACER, B_ACER, C_ACER, D_ACER, E_ACER
```
