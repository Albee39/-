```python
df1 = data1.to_frame(name="LRUN64TTJPA156S")
df2 = data2.to_frame(name="FPCPITOTLZGJPN")
df3 = data3.to_frame(name="DDDM02JPA156NWDB")
```

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_merged=pd.concat([df1,df2,df3],axis=1)
plt.plot(df_merged.index, df_merged["LRUN64TTJPA156S"])
plt.plot(df_merged.index, df_merged["FPCPITOTLZGJPN"])
plt.plot(df_merged.index, df_merged["DDDM02JPA156NWDB"])
```

```python
fig,ax = plt.subplots()
ax.plot(df_merged.index, df_merged['LRUN64TTJPA156S'], color="blue", label='Unemployment Rate')
ax.set_ylabel('Unemployment Rate (%)', color="blue", fontsize=12)
ax2=ax.twinx()
ax2.plot(df_merged.index, df_merged['FPCPITOTLZGJPN'], color="red", label='Inflation Rate')
ax2.set_ylabel('Inflation, consumer prices (%)', color="red", fontsize=12)
ax3=ax.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(df_merged.index, df_merged['DDDM02JPA156NWDB'], color="green", label='Stock Market Traded/GDP')
ax3.set_ylabel('Stock Market Total Value Traded to GDP (%)', color="green", fontsize=12)
lines_labels = [ax.get_legend_handles_labels(), ax2.get_legend_handles_labels(), ax3.get_legend_handles_labels()]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.suptitle('Economic Indicators for Japan', fontsize=16, fontweight='bold')
fig.legend(lines, labels, loc='upper left', bbox_to_anchor=(0.1, 0.95), fontsize=10)
fig.savefig('.jpg',format='jpeg')
```
