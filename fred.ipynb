```python
df1 = data1.to_frame(name='MICH')
df2 = data2.to_frame(name='UNRATE')
```

```python

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_merged=pd.concat([df1,df2],axis=1)
plt.plot(df_merged.index, df_merged['MICH'])
plt.plot(df_merged.index, df_merged['UNRATE'])
```

```python
fig,ax = plt.subplots()
ax.plot(df_merged.index, df_merged['MICH'],color="blue",marker="o")
ax2=ax.twinx()
ax2.plot(df_merged.index, df_merged['UNRATE'],color="red")
fig.savefig('折線圖.jpg',format='jpeg')
```
